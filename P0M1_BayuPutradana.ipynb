#  IV. Data Cleaning

Sebelum melakukan analisis lebih lanjut, langkah awal yang dilakukan adalah proses **data cleaning** pada dataset `shopping_behavior_updated.csv`.  
Tujuannya adalah memastikan data dalam kondisi rapi, konsisten, dan siap dianalisis.

###  Langkah-langkah Data Cleaning
1. **Standardisasi Nama Kolom**  
   - Semua nama kolom diubah ke format `huruf_kecil` (huruf kecil dengan underscore).  
   - Hal ini mempermudah pemanggilan kolom pada Python.  

2. **Menghapus Duplikasi**  
   - Menghapus baris data yang terduplikasi agar tidak memengaruhi hasil analisis.  

3. **Menangani Missing Values**  
   - Kolom bertipe *kategorikal* diisi dengan **modus** (nilai yang paling sering muncul).  
   - Kolom bertipe *numerik* diisi dengan **median** (untuk mengurangi pengaruh outlier).  

4. **Mengubah Tipe Data Kategori Numerik**  
   - Beberapa kolom berisi jawaban *Yes/No* seperti `subscription_status`, `discount_applied`, dan `promo_code_used`.  
   - Agar lebih mudah dianalisis, nilai tersebut diubah ke bentuk numerik:  
     - **Yes â†’ 1**  
     - **No â†’ 0**  
   - Dengan cara ini, data bisa langsung digunakan untuk perhitungan statistik selanjutnya.  

5. **Reset Index**  
   - Index direset agar data rapi setelah pembersihan.  


```python
# ============================================
#  DATA CLEANING
# Dataset: shopping_behavior_updated.csv
# ============================================

import pandas as pd

def clean_data(filepath):
    # --- 0. Load data ---
    df = pd.read_csv(filepath)   # Membaca dataset CSV
    
    # --- 1. Standardisasi nama kolom (snake_case) ---
    df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")  
    # Menghapus spasi, ubah huruf jadi lowercase, dan ganti spasi dengan underscore
    
    # --- 1a. Rename kolom khusus ---
    # Rename kolom purchase_amount_(usd) menjadi purchase_amount_usd untuk konsistensi
    df.rename(columns={'purchase_amount_(usd)': 'purchase_amount_usd'}, inplace=True)
    
    # --- 2. Hapus duplikat (jika ada) ---
    df = df.drop_duplicates()   # Menghapus baris duplikat
    
    # --- 3. Tangani missing values ---
    # Untuk kolom kategorikal â†’ isi dengan modus
    for col in df.select_dtypes(include='object').columns:
        if df[col].isnull().sum() > 0:
            df[col] = df[col].fillna(df[col].mode()[0])   # Isi NaN dengan modus (nilai terbanyak)
    
    # Untuk kolom numerik â†’ isi dengan median
    for col in df.select_dtypes(include=['int64', 'float64']).columns:
        if df[col].isnull().sum() > 0:
            df[col] = df[col].fillna(df[col].median())   # Isi NaN dengan median
    
    # --- 4. Ubah tipe data jika perlu ---
    if 'subscription_status' in df.columns:
        df['subscription_status'] = df['subscription_status'].map({'Yes': 1, 'No': 0})
    
    if 'discount_applied' in df.columns:
        df['discount_applied'] = df['discount_applied'].map({'Yes': 1, 'No': 0})
    
    if 'promo_code_used' in df.columns:
        df['promo_code_used'] = df['promo_code_used'].map({'Yes': 1, 'No': 0})
    
    # --- 5. Reset index (optional) ---
    df = df.reset_index(drop=True)   # Reset index agar urut kembali
    
    return df

# --- Jalankan fungsi cleaning ---
df = clean_data("shopping_behavior_updated.csv")
print(df.head())   # Preview 5 baris pertama
print(df.info())   # Info struktur dataframe

# --- Export Data Bersih ---
def export_clean_data(df, filename="shopping_behaviour_clean.csv"):
    """
    Simpan dataframe yang sudah dibersihkan ke file CSV.
    """
    df.to_csv(filename, index=False)   # Simpan ke CSV tanpa index
    print(f" Data berhasil disimpan ke: {filename}")

# === Panggil fungsi export ===
export_clean_data(df)
```

## Analisis Statistik Deskriptif
Statistik deskriptif digunakan untuk memahami pola dasar data. Fokus pada:
- **Central Tendency (tendensi sentral)** â†’ Mean, Median, Modus.
- **Measure of Variance (penyebaran)** â†’ Variance, Standard Deviation, Range.
- **Outlier Analysis** â†’ deteksi data ekstrem dengan boxplot/IQR. 

```python
# ============================================
#  Statistik Deskriptif untuk Purchase Amount (USD)
# ============================================

import numpy as np
import pandas as pd

# --- 1. Pilih kolom target ---
purchase_col = 'purchase_amount_usd'  # Nama kolom yang dianalisis
purchase = pd.to_numeric(df[purchase_col], errors='coerce').dropna()  # Konversi ke numerik & hapus NaN

# --- 2. Hitung statistik deskriptif dasar ---
mean = purchase.mean()   # Rata-rata
median = purchase.median()   # Nilai tengah
mode = purchase.mode().iat[0] if not purchase.mode().empty else np.nan  # Nilai yang paling sering muncul
variance = purchase.var()   # Varians (tingkat penyebaran data)
std = purchase.std()   # Standar deviasi
min_val = purchase.min()   # Nilai minimum
max_val = purchase.max()   # Nilai maksimum
range_val = max_val - min_val   # Rentang (max - min)

# --- 3. Hitung kuartil & outlier dengan metode IQR ---
Q1 = purchase.quantile(0.25)   # Kuartil 1
Q3 = purchase.quantile(0.75)   # Kuartil 3
IQR = Q3 - Q1   # Interquartile Range
outliers = purchase[(purchase < (Q1 - 1.5 * IQR)) | (purchase > (Q3 + 1.5 * IQR))]  # Deteksi outlier

# --- 4. Tampilkan hasil statistik ---
print(f'Mean: {mean:.2f}')
print(f'Median: {median:.2f}')
print(f'Mode: {mode:.2f}')
print(f'Variance: {variance:.2f}')
print(f'Std: {std:.2f}')
print(f'Range: {range_val:.2f}')
print(f'Jumlah outlier (IQR method): {len(outliers)}')

# --- 5. Fungsi untuk export hasil statistik ke CSV ---
def export_stats_to_csv(filename="purchase_stats.csv"):
    """
    Simpan hasil statistik deskriptif ke file CSV
    untuk pengolahan data lebih lanjut di dashboard.
    """
    stats_dict = {
        "Mean": [mean],
        "Median": [median],
        "Mode": [mode],
        "Variance": [variance],
        "Std": [std],
        "Min": [min_val],
        "Max": [max_val],
        "Range": [range_val],
        "Q1": [Q1],
        "Q3": [Q3],
        "IQR": [IQR],
        "Jumlah Outlier (IQR)": [len(outliers)]
    }
    stats_df = pd.DataFrame(stats_dict)
    stats_df.to_csv(filename, index=False)
    print(f" Statistik deskriptif berhasil disimpan ke: {filename}")

# --- 6. Simpan statistik deskriptif ke file CSV ---
export_stats_to_csv("Statistik Deskriptif.csv")
```

#  Uji Statistik Inferensial: Analisis Lengkap

## 1. Confidence Interval (CI)
Confidence Interval dipakai untuk memperkirakan rata-rata populasi berdasarkan data sampel.  
Misalnya: kita ingin menghitung **95% CI rata-rata purchase_amount_usd**.


```python
# ============================================
#  Confidence Interval 95%
# ============================================

import numpy as np
import scipy.stats as stats
import pandas as pd

# Baca data
df = pd.read_csv("shopping_behaviour_clean.csv")

# Ambil kolom purchase_amount_usd tanpa NA
data = df['purchase_amount_usd'].dropna()

# Hitung statistik dasar
mean = np.mean(data)                 # Rata-rata
std = np.std(data, ddof=1)           # Standar deviasi sampel
n = len(data)                        # Jumlah data

# Tentukan alpha
alpha = 0.05

# Hitung Confidence Interval dengan distribusi t
ci_low, ci_high = stats.t.interval(
    1 - alpha, df=n-1, loc=mean, scale=std/np.sqrt(n)
)

# Cetak hasil dalam format desimal
print(f"Jumlah data: {n}")
print(f"Mean: {mean:.2f}")
print(f"95% Confidence Interval: ({ci_low:.2f}, {ci_high:.2f})")
```

**Kesimpulan:**  

Dengan tingkat keyakinan 95%, rata-rata pembelian diperkirakan berada pada rentang **59.02 â€“ 60.51 USD**.  
Karena intervalnya sempit, perkiraan rata-rata cukup stabil.

---

## 2. Uji Signifikansi (One-Sample t-test)
Untuk menguji apakah rata-rata sampel berbeda signifikan dari nilai tertentu.

```python
# ============================================
#  One-Sample t-Test (Âµ = 60) --> hipotesis nilai rata2 belanja citizen Amerika Serikat sekitar 60 usd
# ============================================

import scipy.stats as stats

def one_sample_ttest(data, mu0=60):   # Uji rata-rata satu sampel terhadap hipotesis harga 60 usd
    data = data.dropna()
    t_stat, p_val = stats.ttest_1samp(data, mu0)
    t_stat, p_val = float(t_stat), float(p_val)
    print(f"T-Statistic: {t_stat:.4f}")
    print(f"P-Value   : {p_val:.4f}")
    return round(t_stat,4), round(p_val,4)

# Jalankan fungsi
one_sample_ttest(df['purchase_amount_usd'], mu0=60)

```

**Interpretasi:**  
- H0: rata-rata = 60  
- H1: rata-rata â‰  60  

**Kesimpulan:**  
Karena *p-value* (0.53) > 0.05, maka **sesuai dengan H0**.  
Artinya, tidak ada bukti signifikan bahwa rata-rata pembelian diluar dari 60 USD sehingga kisaran harga tersebut merupakan harga yang optimal untuk memasarkan produk.

---

## 3. Two-Sample Independent t-test
Menguji apakah dua kelompok independen berbeda signifikan.
Contoh: pria vs wanita.

```python
# ============================================
#  Independent t-Test (Gender)
# ============================================

# Pisahkan data berdasarkan gender
male_purchase = df[df['gender'] == 'Male']['purchase_amount_usd'].dropna()
female_purchase = df[df['gender'] == 'Female']['purchase_amount_usd'].dropna()

# Uji independent t-test (asumsi varians tidak sama)
t_stat_ind, p_val_ind = stats.ttest_ind(male_purchase, female_purchase, equal_var=False)

# Cetak hasil dalam format desimal
print(f"t-statistic: {t_stat_ind:.2f}")
print(f"p-value: {p_val_ind:.4f}")

if p_val_ind < 0.05:
    print("Hasil: Ada perbedaan signifikan pembelian antara pria dan wanita")
else:
    print("Hasil: Tidak ada perbedaan signifikan pembelian antara pria dan wanita")

```

**Interpretasi:**  
- H0: rata-rata pembelian pria = wanita  
- H1: rata-rata pembelian pria â‰  wanita  

**Kesimpulan:**  
Dengan *p-value* (0.377) > 0.05, maka **tidak ada perbedaan signifikan** rata-rata total pembelian antara pria dan wanita.  
Jadi, gender tidak berpengaruh besar terhadap total jumlah belanja.

---

## 4. ANOVA (Analysis of Variance)
Untuk menguji apakah rata-rata lebih dari 2 kelompok berbeda signifikan.
Contoh: pembelian berdasarkan kategori produk.

```python
# ============================================
#  ANOVA (Review Rating ~ Subscription Status)
# ============================================


import scipy.stats as stats

def anova_test(df, cat_col='category', val_col='purchase_amount_usd', n_cat=3):   # Uji perbedaan mean antar kategori
    categories = df[cat_col].unique()[:n_cat]
    groups = [df[df[cat_col]==c][val_col].dropna() for c in categories]
    f_stat, p_val = stats.f_oneway(*groups)
    f_stat, p_val = float(f_stat), float(p_val)
    print(f"F-Statistic: {f_stat:.4f}")
    print(f"P-Value   : {p_val:.4f}")
    print(f"Categories: {list(categories)}")
    return round(f_stat,4), round(p_val,4), list(categories)

# Jalankan fungsi
anova_test(df, 'category', 'purchase_amount_usd', 3)

```

## Interpretasi

1. **Hipotesis ANOVA**  
   - Hâ‚€ (null) : Rata-rata pengeluaran (`purchase_amount_usd`) pada kategori *Clothing*, *Footwear*, dan *Outerwear* **tidak berbeda signifikan**.  
   - Hâ‚ (alternatif) : Ada **perbedaan rata-rata signifikan** minimal di salah satu kategori.  

2. **Keputusan Uji (Î± = 0.05)**  
   - P-Value = 0.1179 > 0.05  
   - â†’ **Sesuai dengan Hipotesis Hâ‚€**  

3. **Kesimpulan**  
   Tidak terdapat perbedaan rata-rata pengeluaran yang signifikan secara statistik antar kategori **Clothing**, **Footwear**, dan **Outerwear** pada tingkat signifikansi 5%.  

4. **Interpretasi Praktis**  
   - Konsumen membelanjakan jumlah yang relatif sama pada ketiga kategori ini.  
   - Perbedaan nilai rata-rata yang ada secara numerik **tidak cukup kuat** untuk dianggap signifikan secara statistik.  
---

## 5. Chi-Square Test
Untuk uji hubungan antara dua variabel kategorikal.
Contoh: Apakah ada hubungan antara gender dan kategori produk yang dibeli.

```python
## Hasil Uji Chi-Square Test (Gender vs Category)


"""
Chi-Square Test (Gender vs Category)
"""
import pandas as pd
import scipy.stats as stats

def chi_square_test(df, col1='gender', col2='category'):   
    # Buat tabel kontingensi
    contingency = pd.crosstab(df[col1], df[col2])
    
    # Hitung Chi-Square
    chi2, p_val, dof, exp = stats.chi2_contingency(contingency)
    
    # Konversi ke float agar tidak keluar np.float64
    chi2, p_val = float(chi2), float(p_val)
    
    # Tampilkan hasil dengan format rapi
    print(f"Chi2 Statistic     : {chi2:.4f}")
    print(f"P-Value           : {p_val:.4f}")
    print(f"Degrees of Freedom: {dof}")
    
    return round(chi2,4), round(p_val,4), dof

# Jalankan fungsi
chi_square_test(df, 'gender', 'category')   # Uji hubungan antara gender & kategori
```

## Hasil Uji Chi-Square Test (Gender vs Category)

### Hasil Uji
- **Chi2 Statistic** : 0.5984  
- **P-Value** : 0.8968  
- **Degrees of Freedom** : 3  

---

### Interpretasi
1. **Hipotesis Chi-Square**  
   - Hâ‚€ (null) : Tidak ada hubungan antara **gender** dan **kategori belanja**.  
   - Hâ‚ (alternatif) : Ada hubungan antara **gender** dan **kategori belanja**.  

2. **Keputusan Uji (Î± = 0.05)**  
   - P-Value = 0.8968 > 0.05  
   - â†’ **Sesuai dengan Hâ‚€**  

3. **Kesimpulan**  
   Tidak terdapat hubungan yang signifikan secara statistik antara **gender** dan **kategori belanja** pada tingkat signifikansi 5%.  
   Nilai Degrees of Freedom (3) menunjukkan jumlah kategori yang dibandingkan setelah dikurangi 1 derajat kebebasan untuk tiap dimensi tabel kontingensi (gender Ã— kategori).  

4. **Interpretasi Praktis**  
   - Pilihan kategori belanja konsumen **tidak dipengaruhi** oleh gender.  
   - Baik laki-laki maupun perempuan memiliki pola distribusi belanja kategori yang relatif sama.  
   - Derajat kebebasan/ degree of freedom (3) berarti uji ini membandingkan variasi distribusi antara **2 baris (gender)** dan **4 kolom (kategori)** sehingga (2-1)Ã—(4-1) = 3.  


---

```python
## Export Hasil Statistik Inferensial ke CSV 


import pandas as pd
import numpy as np
import scipy.stats as stats

def export_inferential_results_clean(df, filename="inferential_statistics_results.csv"):
    results = []

    # 1. Confidence Interval
    data = df['purchase_amount_usd'].dropna()
    mean = float(np.mean(data))  # rata-rata
    std = float(np.std(data, ddof=1))  # standar deviasi sampel
    n = len(data)  # jumlah sampel
    ci = stats.t.interval(0.95, df=n-1, loc=mean, scale=std/np.sqrt(n))  # interval kepercayaan 95%
    results.append({
        "Test": "Confidence Interval",
        "Mean": round(mean,2),
        "Lower CI": round(float(ci[0]),2),
        "Upper CI": round(float(ci[1]),2),
        "Statistic": 0,
        "P-Value": 0,
        "Categories": "None",
        "Degrees of Freedom": 0
    })

    # 2. One-sample t-test (menguji apakah mean sama dengan 60)
    t_stat, p_val = stats.ttest_1samp(data, 60)
    results.append({
        "Test": "One-sample t-test",
        "Mean": 0,
        "Lower CI": 0,
        "Upper CI": 0,
        "Statistic": round(float(t_stat),4),  # nilai t-statistik
        "P-Value": round(float(p_val),4),     # nilai p-value
        "Categories": "None",
        "Degrees of Freedom": 0
    })

    # 3. Two-sample t-test (membandingkan Male vs Female)
    male = df[df['gender']=="Male"]['purchase_amount_usd'].dropna()
    female = df[df['gender']=="Female"]['purchase_amount_usd'].dropna()
    t_stat, p_val = stats.ttest_ind(male, female, equal_var=False)
    results.append({
        "Test": "Two-sample t-test",
        "Mean": 0,
        "Lower CI": 0,
        "Upper CI": 0,
        "Statistic": round(float(t_stat),4),
        "P-Value": round(float(p_val),4),
        "Categories": "None",
        "Degrees of Freedom": 0
    })

    # 4. ANOVA (Analysis of Variance) untuk 3 kategori pertama
    categories = df['category'].unique()[:3]
    groups = [df[df['category']==c]['purchase_amount_usd'].dropna() for c in categories]
    f_stat, p_val = stats.f_oneway(*groups)
    results.append({
        "Test": "ANOVA",
        "Mean": 0,
        "Lower CI": 0,
        "Upper CI": 0,
        "Statistic": round(float(f_stat),4),  # nilai F-statistik
        "P-Value": round(float(p_val),4),     # nilai p-value
        "Categories": ", ".join(categories),  # kategori yang dibandingkan
        "Degrees of Freedom": 0
    })

    # 5. Chi-Square Test (hubungan Gender vs Category)
    contingency = pd.crosstab(df['gender'], df['category'])
    chi2, p_val, dof, exp = stats.chi2_contingency(contingency)
    results.append({
        "Test": "Chi-Square",
        "Mean": 0,
        "Lower CI": 0,
        "Upper CI": 0,
        "Statistic": round(float(chi2),4),   # nilai Chi-square
        "P-Value": round(float(p_val),4),    # nilai p-value
        "Categories": "None",
        "Degrees of Freedom": dof            # degree of freedom hasil uji
    })

    # Buat DataFrame dan isi NaN dengan 0 agar rapi
    results_df = pd.DataFrame(results).fillna(0)

    # Export ke CSV
    results_df.to_csv(filename, index=False)
    print(f"Hasil statistik berhasil diexport ke {filename}")
    return results_df

# Jalankan fungsi
export_inferential_results_clean(df, "Statistik Inferential.csv")


```
